```python
from src.model_io import load_model

model_data = load_model("models/model_v2.npz")

model_data.keys()
```

```python
# Prepare sample data for inference
import pandas as pd

df = pd.read_csv("data/raw/housing.csv")
sample = df.sample(5, random_state=42)
```

```python
# Apply the same feature engineering
from src.feature_engineering import apply_feature_engineering

sample_fe = apply_feature_engineering(sample)

required = model_data["feature_names"]
sample_ordered = sample_fe[required]
```

```python
# Standardize using saved params
mean = model_data["means"]
std = model_data["std"]
mean_s = pd.Series(mean, index=required)
std_s = pd.Series(std, index=required)

sample_scaled = (sample_ordered - mean_s) / std_s
```

```python
# Predict using saved weights
import numpy as np

w = model_data["weights"]
b = model_data["bias"]

y_pred = sample_scaled.to_numpy() @ w + b
y_pred
```
